```python
from elasticsearch.helpers import bulk

mappings = {
        "properties": {
            "parent_category": {"type": "text", "analyzer": "standard"},
            "sun_category": {"type": "text", "analyzer": "standard"},
            "child_category": {"type": "text", "analyzer": "standard"},

    }
}

client.indices.create(index="prod_categories", mappings=mappings)

with open(csv_path) as csv_file:
    reader = csv.DictReader(csv_file)
    counter = 0 

    bulk_data = []
    for row in reader:
        counter += 1
        bulk_data.append(
            {
                "_index": "prod_categories",
                "_id": row['product_id'],
                "_source": {        
                    "parent_category": row["parent_category"],
                    "sun_category": row["sun_category"],
                    "child_category": row["child_category"],
           
                }
            }
        )
    bulk(client, bulk_data)
```
